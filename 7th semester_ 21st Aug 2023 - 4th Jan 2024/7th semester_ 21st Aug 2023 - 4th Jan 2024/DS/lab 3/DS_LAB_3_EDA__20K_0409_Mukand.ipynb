# **Task 3**

```python

data = {'Name': ['C', 'Ali', 'Ahmed', 'Nida', np.nan],
        'Field': ['C', 'E', 'E', 'C', 'C'],
        'Age': [np.nan, np.nan, np.nan, np.nan, np.nan],
        'Marks': [-90, 60, -10, 70, 75]}
df = pd.DataFrame(data)

# Drop the 'Age' column
df = df.drop(columns=['Age'])

# Replace empty strings in the 'Name' column with '---'
df['Name'].replace('', '---', inplace=True)

#  Replace 'C' with 0 and 'E' with 1 in the 'Field' column
df['Field'].replace({'C': 0, 'E': 1}, inplace=True)

#  Replace negative values in the 'Marks' column
Avgmarks = df.loc[df['Marks'] >= 0, 'Marks'].mean()
df['Marks'] = df['Marks'].apply(lambda x: Avgmarks if x < 0 else x)

print(df)
```
