```python
import pandas as pd
from numpy import *
```

## DataSet 1 – 1000 Cameras

Link: https://www.kaggle.com/datasets/crawford/1000-cameras-dataset

```python
# Reading the data into a dataframe

df1 = pd.read_csv("camera_dataset.csv")
df1
```

```python
# dropping non-numeric column

df1.drop(columns = ["Model"], inplace = True)
df1
```

```python
# Dealing with NaN values

for col in df1:
    df1[col].replace({nan: df1[col].mean()}, inplace = True)
```

```python
df1.isna().sum() # Values removed successfully
```

```python
# Making an array

arr1 = df1.to_numpy()
arr1
```

```python
arr1_t = arr1.transpose()
arr1_t
```

```python
product1 = matmul(arr1, arr1_t)
product1
```

```python
eigenvalues1, eigenvectors1 = linalg.eig(product1)
print(eigenvalues1)
print(eigenvectors1)
```

```python
# Remove the eigenvector with lowest eigenvalue

eigenvectors1 = delete(eigenvectors1, argmin(eigenvalues1), axis = 1)
```

```python
print(arr1.shape)
print(eigenvectors1.shape)
```

```python
arr1 = insert(arr1, [12], zeros((1038, 1026)), axis = 1)
```

```python
# Multiplication now possible!

print(arr1.shape)
print(eigenvectors1.shape)
```

```python
PCA1 = matmul(arr1, eigenvectors1)
PCA1
```

```python
U1, S1, V1 = linalg.svd(arr1, full_matrices = True)
```

```python
U1.shape, V1.shape, S1.shape
```

```python
print(U1)
print(V1)
print(S1)
```
